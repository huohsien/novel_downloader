```python
#setup option for chrome profile
chrome_options = Options()
# chrome_options.add_argument("user-data-dir=/Users/huohsien/Library/Application Support/Google/Chrome/Default/")

#start web driver
driver = webdriver.Chrome(ChromeDriverManager().install(), options=chrome_options)
driver.implicitly_wait(60)

wait = WebDriverWait(driver, 30)
```

```python
root = 'https://www.zwxiaoshuo.com/cht/book/7/7181/'
```

```python
driver.get(root)
```

```python
titles = []
hrefs = []
chapters = []

a_elms = driver.find_elements_by_xpath("//strong//a")
for a_elm in a_elms:
    title = a_elm.get_attribute('innerHTML')
    href = a_elm.get_attribute('href')
    if '第' in title and '章' in title:
#         print(title, ' ', href)
        titles.append(title)
        hrefs.append(href)
        chapters.append([title, href])

chapters = np.array(chapters)
```

```python
ch = chapters[:,0].reshape(-1,1)

ch
```

```python
volumes = []
volume = ''
for idx, title in enumerate(chapters[:,0]):
    
    items = title.split(' ')
    if len(items) == 3:
        volume = items[0]
    elif len(items) == 2:
        if items[0].lstrip()[0] == '第':
            print("Miss Chapter title (", title, ").)
        else:
            print("Miss Volume text (", title, "). current volume: ", volume)
            volume = volume + ' ' + title
            print("Fixed volume name: ",volume)
            chapters[idx,0] = volume
            continue
    else:
        print("title: ", title)
        
    if '第' in volume and '章' in volume:
        continue
    if volume == '盜墓筆記柒':
        print("** Modify **")
        volume = '盜墓筆記7'
        chapters[idx,0] = volume
    
    if volume in volumes:
        continue
    #collect unique volume orderly
    volumes.append(volume)
    print("volume: ", volume)
```
